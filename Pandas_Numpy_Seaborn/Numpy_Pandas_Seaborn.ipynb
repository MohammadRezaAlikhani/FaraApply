```python
#import packages
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create numpy arrays
scores = np.random.randint(low=8, high=20, size=5)

# define a list of names
names = ["Ali", "Nika", "Mahsa", "Mohammad", "Nazanin"]
```

```python
# what is data types?
print (f"Type scores: {type(scores)}")
print (f"Type names: {type(names)}")
```

```python
#create table
score_df = pd.DataFrame({"Name": names, "Score": scores})
```

```python
score_df
```

```python
len(score_df)
```

How to add or drop an specific row from Pandas dataframe

```python
score_df.loc[len(score_df)-1]
```

```python
score_df.loc[len(score_df)] = ["Arezoo", 19]
```

```python
score_df.loc[2] = ["Mahnaz", 6.8]
```

```python
score_df.drop(index=3, axis=0, inplace=True)
```

```python
# visualization
plt.figure(figsize=(10,5))

#sns.barplot(data=score_df, x="Name", y="Score")
sns.barplot(x=score_df["Name"], y=score_df["Score"], hue=score_df["Name"])
#plt.bar(x=score_df["Name"], height=score_df["Score"])

plt.title("Student's score")
```

```python
house_df = pd.read_csv("NY-House-Dataset.csv")
house_df.head()
```

```python
house_df[(house_df["TYPE"]=='Condo for sale')]
```

groupby()

```python
new_df = house_df.groupby("TYPE")[["PRICE"]].min().reset_index()
new_df
#<dataframe_name>.groupby(<column for grouping>)<target column>.<aggregation function>
```

```python
plt.figure(figsize=(15,5))

plt.title("Min price for each property group")
sns.barplot(data=new_df, x="TYPE", y="PRICE")
```
