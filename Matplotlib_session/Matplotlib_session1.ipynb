```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dolar.csv")
df.head()
```

What is `Figure` and `Axes` in matplotlib?

`Axes`

`Axis`

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# step 1: figure setting
plt.figure(figsize=(10,5))
sns.set_theme(style='white')

# data visualization
#plt.scatter(df["Date"], df["Close"], label="Close price in scatter format", s=3, c="red")
sns.scatterplot(data=df, x="Date", y="Close", hue="Close")
sns.lineplot(data=df, x="Date", y="High", label="Line plot for High price")


plt.title(label="Random title")
plt.xlabel(xlabel="Date in Georgian format")
plt.ylabel(ylabel="Price in Rials")

plt.grid(visible=False)
plt.legend()

#plt.savefig("test.png")
```
