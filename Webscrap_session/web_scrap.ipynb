# Scrape BAMA.ir website

```python
import requests
```

```python
model = "porsche"
min_price = "3000000000"
max_price = "26000000000"

url = f"https://bama.ir/car/{model}?price={min_price},{max_price}"

header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
```

```python
response = requests.get(url=url, headers=header)
response.status_code
```

```python
bama_soup = BeautifulSoup(response.content, "html.parser")
```

```python
type(bama_soup)
```

```python
# check if we get the correct webpage
with open (file="bama_toyota.html", mode="w", encoding="utf-8") as f:
    f.write(response.text)
```

```python
ad_list = bama_soup.select(selector=".listing")
#ad_list
```

```python
for i in ad_list:
    print(f"Name of the car: {i.select(selector='.bama-ad__title')[0].text.strip()}")
    print(f"Detail of the car: {i.select(selector='.bama-ad__detail-row')[0].text.strip()}")
    print(f"Price of the car: {i.select(selector='.bama-ad__price')[0].text.strip()} Tooman\n")
    print(20*"==")
```

# scrape webpages with dynamic content

```python
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome()
```

```python
model = "porsche"
min_price = "3000000000"
max_price = "26000000000"

url = f"https://bama.ir/car/{model}?price={min_price},{max_price}"

driver.get(url)
```

```python
while True:
    old_height = driver.execute_script("return document.body.scrollHeight")

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    new_height =driver.execute_script("return document.body.scrollHeight")
    time.sleep(600)
    if new_height == old_height:
        break
```

```python
page_source = driver.page_source
```

```python
selenium_soup = BeautifulSoup(page_source, "html.parser")
```

```python
selenium_ad_list = selenium_soup.select(selector=".listing")
#ad_list

for i in selenium_ad_list:
    print(f"Name of the car: {i.select(selector='.bama-ad__title')[0].text.strip()}")
    print(f"Detail of the car: {i.select(selector='.bama-ad__detail-row')[0].text.strip()}")
    print(f"Price of the car: {i.select(selector='.bama-ad__price')[0].text.strip()} Tooman\n")
    print(20*"==")
```

```python
driver.close()
```
